## 1. Exploratory Data Analysis (EDA) and Preprocessing

In this notebook, we'll load the movie dataset, explore the features, look for missing values, and preprocess the text/JSON columns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json

# Set plot style
plt.style.use('ggplot')
sns.set_palette('pastel')

import warnings
warnings.filterwarnings('ignore')
```

### Load the Data

```python
df = pd.read_csv('../data/movie_dataset.csv')
df.head()
```

### Basic Analysis

```python
df.info()
```

```python
df.describe()
```

### Parse JSON Columns (genres and cast)
The genres and cast columns contain stringified JSON. We need to extract them into useful formats.

```python
def extract_names(item_str):
    if pd.isna(item_str):
        return []
    try:
        items = json.loads(item_str)
        return [item['name'] for item in items]
    except:
        return []

df['genres_list'] = df['genres'].apply(extract_names)
df['cast_list'] = df['cast'].apply(extract_names)

df.head(3)
```

### Visualizations

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='budget', y='revenue', data=df, alpha=0.5)
plt.title('Budget vs Revenue')
plt.xlabel('Budget ($)')
plt.ylabel('Revenue ($)')
plt.tight_layout()
plt.show()
```

### Save Preprocessed Data

```python
df.to_csv('../data/cleaned_movie_dataset.csv', index=False)
print("Saved preprocessed data to cleaned_movie_dataset.csv")
```
